```python
df_kosis.head(1)
```

```python
df = df_kosis.melt(id_vars="시군구별")
df.head()
# variable에 시군구별 값이 들어감
```

```python
# 이런상황에서 데이터 전처리를 통하여 연도별 월별등으로 분석가능!
df["시군구별"].unique()
```

```python
df[df["시군구별"]=="시군구별"]
```

```python
df[df["시군구별"]!="시군구별"]
# 시군구별안의 시군구별과 다른 데이터들
```

```python
print(df.shape)
df = df[df["시군구별"]!="시군구별"].copy() # df 전처리
df.shape # 시군구별 데이터들을 사라지게 정리함
```

```python
df["시군구별"].unique()
```

```python
df.head()
```

```python
df["연도"]=df["variable"].str.split(".", expand=True)[0]
df["월"]=df["variable"].str.split(".", expand=True)[1]
df["성별"]=df["variable"].str.split(".", expand=True)[2]
df.head()
# 문자열을 . 기준으로 자르고 expand=True 하면 데이터 프레임형식으로 가져온다
# 1번은 남성 2번은 여성
# 연도 =0 월 =1 성별 =2 에 있음
```

```python
df.tail()
```

```python
df["성별"].unique()
```

```python
df["성별"].nunique()
```

```python
df["성별"] = df["성별"].fillna("전체")
df["성별"].unique()
# null 값을 전체로 바꿔줌
```

```python
df["성별"].replace("1","남자").replace("2","여자")
df["성별"] = df["성별"].replace("1","남자").replace("2","여자")
df["성별"].unique() 
# 1,2과 남자 여자로 안바뀌어 있음 꼭 다시 할당을 해야함
```

```python
df["성별"].value_counts()
# 시리즈 데이터만 셀수 있고 한개의 데이터의 빈도수를 세고 싶을 때 사용
```

```python
df.head()
```

```python
df.tail()
```

```python
df.rename(columns={"variable":"기간","value":"출생아수"})
```

```python
df= df.rename(columns={"variable":"기간","value":"출생아수"})
```

```python
df.head()
```

```python
df.info()
```

```python
import numpy as np
df["출생아수"]= df["출생아수"].replace("-",np.nan)
# str형식으로 int 타입으로 바뀌어줘야하는데 "-" 가 있어서 오류가 남
# nan으로 바뀌어 주면 float타입이기 때문에
# nan 대신 0으로 하면 평균값이 안나옴.
df["출생아수"] = df["출생아수"].astype(float)
df["출생아수"].describe()
```

```python
df_all = df[(df["시군구별"]=="전국") & (df["성별"]=="전체")] 
df_all.head(2)
# 블리언 인덱싱?
# 전국 기준으로 전체 출생아수를 가져온다.
```

```python
df_all = df_all[["연도","월","출생아수"]].copy()
df_all
# 필요 데이터만 뽑아옴
```

```python
import matplotlib.pyplot as plt
plt.rc("font",family="Malgun Gothic")
# 폰트 설정
```

## 지역별 출생아수 분석

```python
df_local = df[df["시군구별"] != "전국"].copy()
df_local.head()
# df_local를 수정할 것이면 copy사용 아닐경우 copy 안써도된다.
```

```python
sns.pointplot(data=df_local, x= "연도",y="출생아수")
# 평균이 아직 제대로 인것은 아님성별이 전체가 들어가 있어서 전체를 빼거나 전체만하거나 해야함
```

```python
plt.figure(figsize=(15,4))
sns.pointplot(data=df_local, x= "연도",y="출생아수",hue="성별")
```

```python
df_local_all = df_local[df_local["성별"]=="전체"]
df_local_all
```

```python
plt.figure(figsize=(15,4))
sns.pointplot(data=df_local_all, x= "연도",y="출생아수",hue="시군구별")
# 성별의 전체만 빼서 시군구별 그래프로 시각화
# 시군구별로 표시되어 있는 박스가 보기 힘드니 뺄 예정
```

```python
plt.figure(figsize=(15,4))
sns.pointplot(data=df_local_all, x= "연도",y="출생아수",hue="시군구별", ci = None)
plt.legend(loc='center right', bbox_to_anchor=(1.15,0.5), ncol=1)
# bbox위 숫자로 박스의 위치조정!
```

```python
# 특정 시만 가져오고 싶을 경우
df_local_2 = df_local_all[df_local_all["시군구별"].isin(["서울특별시","경기도","세종특별자치시"])]
```

```python
df_local_2.head()
```

```python
plt.figure(figsize=(15,4))
sns.pointplot(data=df_local_2, x= "연도",y="출생아수", ci=None, hue="시군구별")
```
