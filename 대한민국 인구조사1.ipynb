### 위키피디아 그래프 처럼 실습하기

### Pandas를 통해서 그려보기

```python
df[["연도 (년)","출생자수(명)","사망자수(명)"]]
```

```python
df_pop = df[["연도 (년)","출생자수(명)","사망자수(명)"]]
df_pop = df_pop.set_index("연도 (년)") 
df_pop.head()
# 연도가 x축에 오지 않으므로 조정이 필요.
# 연도를 index값으로 지정해야 x축에 표시가능.
# set_index를 통해 지정 가능
```

```python
df_pop.plot(figsize =(15,4)) 
# plot() 안에 figsize로 크기조정 가능
```

```python
# 최근 50년만 그리고 싶은 경우
```

```python
df_pop[-50:].plot(figsize =(15,4))
# 슬라이싱을 이용해서 그림
```

```python
plt.figure(figsize=(15,8))
plt.xticks(rotation=60) 
sns.pointplot(data=df[-50:], x="연도 (년)", y="출생자수(명)")
sns.pointplot(data=df[-50:], x="연도 (년)", y="사망자수(명)", color="orange")
```
