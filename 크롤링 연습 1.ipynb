```python
from bs4 import BeautifulSoup
import requests as req
import pandas as pd
import urllib.request
```

```python
url ="http://comp.fnguide.com/SVO2/ASP/SVD_Finance.asp?pGB=1&gicode=A005930"
response = req.get(url)
# html = urllib.request.urlopen(url).read()
response
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
# 테이블 찾기
tabletag = soup.find(name="div",
                     attrs={"class":"um_table"})
tabletag
```

```python
# 테이블정보 수집하기
# tbody의 tr 태그를 모두 찾는다. 반복으로 찾기
rows = tabletag.find("tbody").findAll("tr")
rows[78]
```

```python
columns = rows[0].findAll(name="td")
```

```python
rows[1].findAll(name="td")
```

```python
columnslen = len(columns)
```

```python
# 값을 가져올 수 있음.
columnlist.append(columns[0].text)
columnlist.append(columns[1].text)
columnlist
```

```python
# 행 전체를 저장할 리스트
rowlist=[]
# 행별 컬럼값을 저장할 리스트
columnlist=[] 

for i in range(0,len(rows)): 
    columns = rows[i].findAll("td") # 행을 반복하면서 td를 모두 찾는다.
    columnslen = len(columns)
    for j in range(0,columnslen): # columnslen의 길이만큼.
        columnlist.append(columns[j].text) # columnslist td값을 다 추가시켜줌
        
    rowlist.append(columnlist) # 전체데이터 수집
    columnlist=[] # 행별 컬럼 데이터 수집 후 컬럼리스트 초기화
```

```python
# 코딩 다른방법
for eachRow in rows: 
    columns = eachRow.findAll("td") # 행을 반복하면서 td를 모두 찾는다.
    columnslen = len(columns)
    for eachColumn in columns: # columnslen의 길이만큼.
        columnlist.append(eachColumn.text) # columnslist td값을 다 추가시켜줌
        
    rowlist.append(columnlist) # 전체데이터 수집
    columnlist=[] # 행별 컬럼 데이터 수집 후 
```
